## 2.2 Finding out if there are any tweets having high ‘favoriteCount’ count and high ‘retweetCount’.
- This helps us find the most influencial tweets in the dataset that majority people also engaged with.

If a tweet has a high favourite count and high retweet count that tells us that it hasnt just garnered most support but also been the basis of major discussion. This can help us identify which tweets helped shape the movement and its associated sentiments and how big of a role they played in molding the public sentiment.

```python
df['combinedCount'] = df['favoriteCount'] + df['retweetCount']
max_combined_row = df[df['combinedCount'] == df['combinedCount'].max()]
tweet_text = max_combined_row['text'].values[0]
favorite_count = max_combined_row['favoriteCount'].values[0]
retweet_count = max_combined_row['retweetCount'].values[0]

print(f"The tweet with the highest favorite and retweet counts:")
print(f"Tweet: {tweet_text}")
print(f"Favorite Count: {favorite_count}")
print(f"Retweet Count: {retweet_count}")

df.drop('combinedCount', axis=1, inplace=True)
```

## 2.3 Analyzing the content of popular tweets like keywords and topics.
- this helps us to better understand the natue of the data and what is being discussed and highlighted in the tweets

Keywords are essential for analysing any text information. Here keywords that are used frequently in the most popular tweets can help us understand the topics that shaped the movement and were widely discussed.

Here we have made DataFreme of the top 10000 tweets with high retweet and favourite count and then have used that along with the entire dataset to find the keywords and see the differences.

```python
#creating DataFrames with top 10000 most popular tweets
top_10000_retweets = df.nlargest(100000, 'retweetCount')
top_10000_fav = df.nlargest(100000, 'favoriteCount')
top_10000_retweets = top_10000_retweets.sort_values(by='retweetCount', ascending=False)
top_10000_fav = top_10000_fav.sort_values(by='favoriteCount', ascending=False)
```

```python
retweets_tweet_text = set(top_10000_retweets['text'])
fav_tweet_text = set(top_10000_fav['text'])
common_tweet_text = retweets_tweet_text.intersection(fav_tweet_text)

common_tweets_df = top_10000_retweets[top_10000_retweets['text'].isin(common_tweet_text)]

common_tweets_df= common_tweets_df.sort_values(by='retweetCount', ascending=False)
common_tweets_df=common_tweets_df[:10000]
common_tweets_df
```

```python
import pandas as pd
from collections import Counter
import re

def extract_keywords(text):
    if isinstance(text, str):
        words = re.findall(r'\w+', text.lower())
        return words
    else:
        return []

all_keywords = df['lemmatized_tweets'].apply(extract_keywords).sum()
keyword_counts = Counter(all_keywords)
most_common_keywords = keyword_counts.most_common(10)
for keyword, count in most_common_keywords:
    print(f"Keyword: {keyword}, Frequency: {count}")
```

From the above results we can infer that keywords from the entire dataset are genralised including words like woman, misconduct, assualt the keywords from the most popular tweets mention names of people and words such as allegation, america, president. # 2.4 Investigating the timing of popular tweets.
- analysing the timings of key tweets and the reaction to them is crucial for understanding the sentiments and how they changed with time

Through the following graph we can clearly visualise the timeline of the popular tweets and how they have imapcted the conversation around #metoo

```python
df_sorted_rt = df.sort_values(by='favoriteCount', ascending=False)
top_100_rt = df_sorted_rt.head(1000)
top_100_rt['created'] = pd.to_datetime(top_100_rt['created'])
pink_color = "#FF6B6B"

fig = px.scatter(
    top_100_rt,
    x='created',
    y='favoriteCount',
    title='Timeline of the 1000 Most Favourited Tweets',
    labels={'created': 'Date and Time', 'favoriteCount': 'Favourite Count'},
    color_discrete_sequence=[pink_color]
)

fig.update_xaxes(title_text='Date and Time')
fig.update_yaxes(title_text='Favourite Count')

fig.update_xaxes(
    tickmode='linear',
    dtick='D1',  # Show dates at daily intervals
    tickformat='%Y-%m-%d',  # Format the date as 'YYYY-MM-DD'
)
fig.show()
```

# 2.5 Examining the twitter users associated with popular tweets.
- This helps us understand who the key players were in this movement

Here firstly we have found the people who have th emost number of tweets in the entire dataset, in a list of the top 10000 retweeted tweets and in a list of the top 10000 popular tweets.

This helps us identify the key Players in the movement who have garnered the most discussion on their tweets and how active they were in the movement.

```python
import pandas as pd
from collections import Counter

df = df.dropna(subset=['screenName'])
mention_counts = Counter(df['screenName'])
most_popular_people = mention_counts.most_common(10)  # Get the top 10 people tweeting the most

print("People with the most tweets:")
for person, count in most_popular_people:
    print(f"{person}: {count} Tweets")
```

```python
top_10000_retweets = df.nlargest(10000, 'retweetCount')
top_10000_fav = top_10000_fav.dropna(subset=['screenName'])
mention_counts_1= Counter(top_10000_fav['screenName'])
most_popular_people_1= mention_counts_1.most_common(10)  # Get the top 10 people tweeting the most

print("People with the most tweets in top 10000 most retweeted Tweets:")
for person, count in most_popular_people_1:
    print(f"{person}: {count} Tweets")
```
